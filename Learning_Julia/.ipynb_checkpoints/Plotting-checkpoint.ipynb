# Plotting

## Basics
There are a few different ways to plot in Julia (including calling PyPlot).
Here we'll look at using `Plots.jl`

```julia
using Plots
```

One of the advantages to `Plots.jl` is that it allows you to seamlessly change backends. In this notebook, we'll try out the `gr()` and `plotlyjs()` backends.

First let's generate some artificial data to plot.

```julia
x = -3:0.1:3
f(x) = x^2

y = f.(x)
```

Now we load the GR backend.

```julia
gr()
```

```julia
plot(x, y, label = "line")
scatter!(x, y, label = "points")
```

Notice that the `scatter!` function has an `!`, meaning that it added to the graph without tking away what was already there. Take a look what happens if you remove the `!` below.

```julia
plot(x, y, label = "line")
scatter(x, y, label = "points")
```

Without changing syntax, we can try this with the plotlyjs() backend. Note: in order to use plotlyjs, I had to use this command in the Julia terminal: `Pkg.add("PlotlyJS"); Pkg.add("ORCA")`

```julia
plotlyjs()
```

```julia
plot(x, y, label = "line")
scatter!(x, y, label = "points")
```

Something cool about plotlyjs is that you can run your cursor over your data and see the points.

## Getting slightly fancier.
Let's look at pirates and temperature.

```julia
global_temp = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
num_pirates = [45000, 20000, 15000, 5000, 400, 17]
```

```julia
#First plot the data
plot(num_pirates, global_temp, legend = false)
scatter!(num_pirates, global_temp, legend = false)
```

You can reverse the x axis using `xflip!()`
To add labels and titles, use: `xlabel!`, `ylabel!`, and `title!`.

```julia
xflip!()
xlabel!("Approximate # of Pirates")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
```

Note that we flipped the x-axis so that from left to right time would be going forwards instead of backwards.

To create a plot with subplots, name the subplots then wrap those names, along with layout commands, in another `plot` call.

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```
